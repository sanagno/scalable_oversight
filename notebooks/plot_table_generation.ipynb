```python
def show_text(text):
    return text.replace("^", "\^{}").encode("ascii", "ignore").decode("utf-8")


advocate_model = "Llama-2-70b-chat"

for dataset in ["openbookqa", "boolq", "piqa", "gpqa"]:
    advocate_data_folder = os.path.join("data/advocate_data", dataset, advocate_model)

    examples = []
    for advocate_level in LEVELS:
        advocate_dataset = load_pickle(
            os.path.join(advocate_data_folder, f"generations_{advocate_level}.pkl")
        )
        examples.append(advocate_dataset[0])
        # dataset_raw = get_dataset(Namespace(base_data_folder="data"), dataset, None, is_advocate=True, )

    print("\\begin{longtable}{| p{.20\\textwidth} | p{.80\\textwidth} |} ")
    print("\\toprule")
    print("Role & Content \\\\")
    print("\\midrule")

    for idx in range(len(examples)):
        print(
            "System &",
            show_text(examples[idx]["conversation_history"][0]["content"]),
            " \\\\",
        )
        print(
            "User &",
            show_text(examples[idx]["conversation_history"][1]["content"]),
            " \\\\",
        )
        print("Assistant &", show_text(examples[idx]["generation"]), " \\\\")

        if idx < len(examples) - 1:
            print("\midrule")

    print("\\bottomrule")
    print("\\caption{" + DATASET_NAMES[dataset] + " explanation examples.}")
    print("\\label{tab:" + dataset + "_examples}")
    print("\\end{longtable}")

    print()
    print()
    print()
    print()
```
